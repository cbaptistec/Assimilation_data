On considère en premier lieu deux variables aléatoires gaussiennes $T_1$ et $T_2$ de même moyenne $T_t$ et de variances $\sigma_1^2$ et $\sigma_2^2$, dont les densités de probabilité sont données par:

$$f_i(T)=\frac{1}{\sqrt{2 \pi \sigma_i^2}} \exp \left(\frac{(T-T_t)^2}{2 \sigma_i^2}\right) \qquad \text{pour }i \in \{1,2\}$$

Pour la suite, on prend $T_t=60$, $\sigma_1=5$ et $\sigma_2=4$.
Le programme fait ensuite un estimation optimale de $T_t$ avec:

$$T_a=\frac{C_1T_1+C_2T_2}{C_1+C_2} \qquad \text{où} \quad C_i=\frac{1}{\sigma_i^2}$$

Pour savoir si notre estimation est bonne, on veut savoir son erreur est inférieur à l'erreur obtenue avec une seule variable aléatoire. On cherche donc à évaluer $\rho=P(|T_a-T_t|<|T_1-T_t|)$. Si cette probabilité est supérieur à 0.5, alors on a fait un progrès.

```python


# Parameters
Tt=60 # True state
sig1=4 ; C1=1/sig1**2 # T1
sig2=5 ; C2=1/sig2**2 # T2
Ca=C1+C2; siga=np.sqrt(1/Ca) # Ta

def fT(T,Tt,sig):
    # Gaussian probability density function 
    K=1/(np.sqrt(2*np.pi)*sig)
    f = K*np.exp(-(T-Tt)**2/(2*sig**2))
    return f,K

def val(Tt,sig,col) :
    # Ploting the probability density functions
    T = np.linspace(Tt-3*sig, Tt+3*sig, 100)
    f,K=fT(T,Tt,sig)
    plt.plot(T, f, color=col, linewidth=3)
    # Random sampling

    #Tr = np.random.normal(Tt,sig,20)
    #fR,KR=fT(Tr,Tt,sig)
    #plt.scatter(Tr,fR,marker='o',linewidth=3,color=col)

    plt.axhline(0)
    plt.axvline(Tt)
    plt.title('Probability density functions',fontsize=16)
    plt.xlabel('$T$',fontsize=32 )
    plt.xticks(fontsize=16)
    plt.yticks(fontsize=16)
    plt.ylabel('$f_{T}$',fontsize=32)
    
 

def histo(i) :
    # Ploting random draws on a line    
    plt.figure()
    val(Tt,sig1,'black');
    val(Tt,sig2,'green')
    val(Tt,siga,'red')
    # one draw
    T1=np.random.normal(Tt,sig1)
    T2=np.random.normal(Tt,sig2)
    Ta=(C1*T1+C2*T2)/Ca
    plt.scatter(Tt,0,marker='^',color='blue',s=200)
    plt.scatter(T1,0,marker='^',color='black',s=200)
    plt.scatter(T2,0,marker='^',color='green',s=200)
    plt.scatter(Ta,0,marker='^',color='red',s=200)

def improve(Nc):
    # Counting if T2 improves the analysis
    count=0
    for i in range(Nc) :
        T1=np.random.normal(Tt,sig1)
        T2=np.random.normal(Tt,sig2)
        Ta=(C1*T1+C2*T2)/Ca      
        if abs(Ta-Tt) < abs(T1-Tt) :
            count=count+1
    
    rho=count/Nc
    return rho




```

On se propose de vérifier l'amélioration de l'estimation de $T_t$ en ne considérant plus deux mais $N_T$ variables aléatoires gaussiennes de variances $\sigma_i^2$. Dans le cas génaral, on a:

$$T_a=\frac{\displaystyle\sum_{i=1}^{N_T}C_i T_i}{\displaystyle\sum_{i=1}^{N_T}C_i } \qquad \text{avec} \quad C_i=\frac{1}{\sigma_i^2}$$


Dans notres cas pour $i \ge 2$, on a pris $\sigma_i=\sigma_2=5$. On trace ensuite le taux d'amélioration de la prédiction par rapport au résultat de la première variable aléatoire. On a choisit de faire $R=1000$ réalisations pour chaque cas.
On a tracé $1-\rho$ en fonction du nombre de variables aléatoires considérées pour voir comment le taux d'amélioration s'approche de 1. 

```python
#Amélioration de l'estimation de T_t avec NT variables aléatoires

def improveNT(Nc,NT):
    # Counting if T2 improves the analysis
    count=0
    for i in range(Nc) :
        T=np.zeros(NT)
        T[0]=np.random.normal(Tt,sig1)
        for j in range(1,NT):
          T[j]=np.random.normal(Tt,sig2)
        Ta=(C1*T[0]+C2*np.sum(T[1:]))/(C1+(NT-1)*C2)     
        if abs(Ta-Tt) < abs(T[0]-Tt) :
            count=count+1
    
    rho=count/Nc
    return rho

# Counting the improvements with NT=3 random variables
if True :
    rho=improveNT(1000,3)
    print('=================================')
    print('Rate of improvemennts with 3 random variables rho=',rho)
    print('=================================')




# Counting the improvements with NT=10 random variables
if True :
    rho=improveNT(1000,10)
    print('=================================')
    print('Rate of improvemennts with 10 random variables rho=',rho)
    print('=================================')





#Plot rate of improvement with NT:

NT_list=np.array([2**i for i in range(1,14)])
rho=[]

for i in NT_list:
  rho.append(improveNT(1000,i))

rho=np.array(rho)
plt.figure()
plt.axhline(0)
plt.axvline(0)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.plot(NT_list,1-rho,color='red')

plt.title('Improvement rate',fontsize=12)
plt.xlabel('Nombre de variables aléatoires',fontsize=12 )
plt.xscale('log')
plt.ylabel('1-rho',fontsize=16)
plt.yscale('log')
plt.savefig('rho.pdf')
plt.grid()
plt.show(); plt.close()
```
